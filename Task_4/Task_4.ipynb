# To Explore Decision Tree Algorithm

```python
# importing libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn import tree

```

```python
plt.figure(figsize=(15,10))
plt.subplot(2,2,1)
sns.violinplot(x='Species',y='PetalLengthCm',data=data)
plt.subplot(2,2,2)
sns.violinplot(x='Species',y='PetalWidthCm',data=data)
plt.subplot(2,2,3)
sns.violinplot(x='Species',y='SepalLengthCm',data=data)
plt.subplot(2,2,4)
sns.violinplot(x='Species',y='SepalWidthCm',data=data)
```

```python
#separating dataset
# x = data[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']].values
# y = data['Species'].values
x = data.iloc[:,1:5].values
y = data.iloc[:,-1].values
```

```python
# defining decision tree algorithm

classify = DecisionTreeClassifier(criterion = 'entropy',random_state = 0)
classify.fit(x,y)
```

```python
# spliiting data into test and train
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.25,random_state = 0)
```

```python
# Feature scaling

from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
# training decision tree model on training data set
classify.fit(x_train,y_train)
```

```python
# prediction of test results

y_pred = classify.predict(x_test)
```

```python
# showing actual and predicted values

tr = pd.DataFrame({'Actual': y_test,'Predicted': y_pred})
tr.head()
```

```python
tr.tail()
```

```python
#Accuracy of Decision Tree

a = accuracy_score(y_pred,y_test)
print(f'accuracy score for decision tree is : {a}')
```

#   visualizing the Decision Tree

```python
from sklearn.externals.six import StringIO
from IPython.display import Image
from sklearn.tree import export_graphviz
import pydotplus
import os

```

```python
dt = StringIO()

class_names =['iris-setosa','iris-versicolor','iris-virginica']
feature_names = list(data.columns[:4])

export_graphviz(classify,out_file = dt,feature_names = feature_names,
               filled = True,rounded=True,special_characters = True)

graph = pydotplus.graph_from_dot_data(dt.getvalue())
Image(graph.create_png())
```
